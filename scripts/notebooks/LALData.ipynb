```python
#read teams csv file
teams_df = pd.read_csv('teams.csv')

teams_df.head()
```

```python
# team of choice
my_team = 'LAL'

team_id = teams_df.loc[teams_df['ABBREVIATION'] == my_team].iloc[0]['TEAM_ID']
print("team_id",team_id)
```

```python
data = games_df.loc[(games_df['TEAM_ID_home']==team_id) | (games_df['TEAM_ID_away']==team_id)]
data.head()
```

```python
data.info()
```

```python
#create a list for team performance over the past 7 games to analyze team strength
teamPerformance = []
for ind in data.index:
   date = data['GAME_DATE_EST'][ind]
   recent_games = games_df[games_df['GAME_DATE_EST'] < date][(games_df['HOME_TEAM_ID'] == team_id) | (games_df['VISITOR_TEAM_ID'] == team_id)].sort_values(by = 'GAME_DATE_EST').tail(7)

  # print(recent-games)

  #calculate the average points scored
   average_points = 0
   for index in recent_games.index:
    if games_df['HOME_TEAM_ID'][ind] == team_id:
      average_points += (games_df['PTS_home'][ind])
    else:
      average_points +=(games_df['PTS_away'][ind])

  # print (points)
   average_points = average_points/7
   teamPerformance.append(average_points)

print(teamPerformance)
```

```python
#create a list for opposition performance over the past 7 games to analyze team strength
opPerformance = []
for ind in data.index:
   # get the opposition team
   if data['HOME_TEAM_ID'][ind]== team_id:
    opp = data['TEAM_ID_away'][ind]
   else:
    opp = data['TEAM_ID_home'][ind]
   date = data['GAME_DATE_EST'][ind]
   recent_games = games_df[games_df['GAME_DATE_EST'] < date][(games_df['HOME_TEAM_ID'] == opp) | (games_df['VISITOR_TEAM_ID'] == opp)].sort_values(by = 'GAME_DATE_EST').tail(7)

  # print(recent-games)

  #calculate the average points scored
   average_points = 0
   for index in recent_games.index:
    if games_df['HOME_TEAM_ID'][ind] == opp:
      average_points += (games_df['PTS_home'][ind])
    else:
      average_points +=(games_df['PTS_away'][ind])

  # print (points)
   average_points = average_points/7
   opPerformance.append(average_points)

print(opPerformance)
```

```python
# Add a new binary column named 'HOME' to data
data['HOME'] = [1 if x == team_id else 0 for x in data['HOME_TEAM_ID']]
data.head()
```

```python
#create a list to hold games won by the team
team_wins=[]

won = 0
for ind in data.index:
  if (data['HOME'][ind] == 1 and data['HOME_TEAM_WINS'][ind] == 1) or (data['HOME'][ind] == 0 and data['HOME_TEAM_WINS'][ind] == 0):
    won = 1
  else:
    won = 0

  team_wins.append(won)

print(team_wins)
```

```python
# add a new column for team performance over the past 7 games
data['TEAM_PERF'] = teamPerformance
```

```python
# add a new column for opposition performance over the past 7 games
data['OPP_PERF'] = opPerformance
```

```python
# add a new column for team win after each game
data['TEAM_WINS'] = team_wins
```
